```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
df.drop("Cabin",axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.countplot(x='Sex', data=df)
plt.title('Distribution of Sex')
plt.savefig("Distribution of Sex.png", dpi=300, bbox_inches="tight")
plt.show()

sns.countplot(x='Pclass', data=df)
plt.title('Distribution of Passenger Class')
plt.savefig("Distribution of Passenger Class.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
sns.histplot(df['Age'].dropna(), kde=True) 
plt.title('Distribution of Age')
plt.savefig("Distribution of Age.png", dpi=300, bbox_inches="tight")
plt.show()

sns.boxplot(y='Fare', data=df)
plt.title('Distribution of Fare')
plt.savefig("Distribution of Fare.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Sex')
plt.ylabel('Survival Rate')
plt.savefig("Survival Rate by Sex.png", dpi=300, bbox_inches="tight")
plt.show()

sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.savefig("Survival Rate by Passenger Class.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
sns.histplot(data=df, x='Age', hue='Survived', kde=True, multiple="stack")
plt.title('Age Distribution by Survival Status')
plt.savefig("Age Distribution by Survival Status.png", dpi=300, bbox_inches="tight")
plt.show()

sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Distribution by Survival Status')
plt.savefig("Fare Distribution by Survival Status.png", dpi=300, bbox_inches="tight")
plt.show()
```
